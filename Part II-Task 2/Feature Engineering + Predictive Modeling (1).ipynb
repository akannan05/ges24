```python
y = df['label_y']
```

```python
y = y.replace(-1, 0)
```

```python
y.value_counts()
```

```python
X = df.drop('label_y', axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42)
```

```python
X_train
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=100)
rf_model.fit(X_train, y_train)
```

```python
predictions = rf_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(predictions, y_test))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
conf_matrix_one = confusion_matrix(y_test, predictions)
```

```python
ConfusionMatrixDisplay(conf_matrix_one).plot()
plt.title("Confusion Matrix (1 = Potential Customer, 0 = Not a Potential Customer)")
```

```python
y_probs = rf_model.predict_proba(X_test)[:, 1]
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.metrics import precision_recall_curve
precision, recall, thresholds = precision_recall_curve(y_test, y_probs)
fig, ax = plt.subplots()
ax.plot(recall, precision, color='purple')

ax.set_title('Precision-Recall Curve')
ax.set_ylabel('Precision')
ax.set_xlabel('Recall')

#display plot
plt.show()
```

```python
from sklearn.metrics import roc_curve
from sklearn import metrics
fpr, tpr, _ = metrics.roc_curve(y_test, y_probs)
roc_auc = metrics.auc(fpr, tpr)
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'AUC = {roc_auc:.2f}')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```

```python
from imblearn.over_sampling import RandomOverSampler
from collections import Counter
oversample = RandomOverSampler(sampling_strategy='minority')
X_over, y_over = oversample.fit_resample(X, y)
print("Oversampled class distribution:", Counter(y_over))
```

```python
X_train_over, X_test_over, y_train_over, y_test_over = train_test_split(
    X_over, y_over, test_size=0.25, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model_over = RandomForestClassifier(n_estimators=100)
rf_model_over.fit(X_train_over, y_train_over)
```

```python
predictions_over = rf_model_over.predict(X_test_over)
```

```python
final_predictions = rf_model_over.predict(X_test)
```

```python
print(classification_report(predictions_over, y_test_over))
```

```python
print(classification_report(final_predictions, y_test))
```

```python
conf_matrix_two = confusion_matrix(y_test, final_predictions)
```

```python
ConfusionMatrixDisplay(conf_matrix_two).plot()
plt.title("Confusion Matrix (1 = Potential Customer, 0 = Not a Potential Customer)")
```

```python
y_score_over = rf_model_over.predict_proba(X_test)[:, 1]
```

```python
from sklearn.metrics import precision_recall_curve
precision, recall, thresholds = precision_recall_curve(y_test, y_score_over)
fig, ax = plt.subplots()
ax.plot(recall, precision, color='purple')

ax.set_title('Precision-Recall Curve')
ax.set_ylabel('Precision')
ax.set_xlabel('Recall')

#display plot
plt.show()
```

```python
from sklearn.metrics import roc_curve
from sklearn import metrics
fpr, tpr, _ = metrics.roc_curve(y_test, y_score_over)
roc_auc = metrics.auc(fpr, tpr)
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'AUC = {roc_auc:.2f}')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```
